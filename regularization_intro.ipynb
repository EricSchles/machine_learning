# An Introduction to Regularization

The goal of this notebook is to introduce readers to regularization techniques.  Before we begin walking through these techniques, let's start with a brief definition of regularization:

_Definition_ **Regularization** - A technique to reduce how much noise a model learns when training.

_Definition_ **Model Noise** - Random information that is not consistent across samples from a population and adds no semantic explanatory power to the variance in Y caused by variance in X.

Given these two definitions, we can understand regularization for what it is, a technique that attempts to reduce learning models that overfit our data.  If our model does overfit our data, it will learn things from the data that are actually random or particular to this sample of data, rather than being a feature of the overall population that the data was drawn from.

_Example_ **How overfitting can arise**

Suppose you have the simple task of trying to identify a wolf from a dog.  Further suppose you want to use a statistical model to do this, because trying to write a program yourself to do it, would be very, very hard.  Further suppose you end up creating such a model that has high accuracy.  The first question you should ask, can I trust my results?  One of the line items in the checklist you should be going over, when trying to answer this question is, is there overfitting in my model?

So how do you go about doing this?  Well one way is to make use of **L**ocal **I**nterpretable **M**odel-agnostic **E**xplanations (LIME).  The LIME package can check for overfitting.  The following picture is taken from [this article](https://hackernoon.com/dogs-wolves-data-science-and-why-machines-must-learn-like-humans-do-41c43bc7f982).  



As you can see from the picture above, the reason the classifier learned that it was a wolf and not a husky is because of the snow in the picture.  This means there is overfitting in the model, because the model is learning whether to classify based on random information that does not have to do with the variables of interest - IE the features of the animal in the picture.  

So how might we deal with this?

One possible method is, we could first have the code look for shapes of dogs or wolfs, because they are generally similar and then only consider pixels within a bounding box around the animal in question.  It turns out this is the essence of regularization - cutting out the noise.  In this example we do it literally.

## A Review of Ordinary Least Squares

The general premise behind ordinary least squares is you start with some data:

```python
import pandas as pd
import random
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return (x*27) + random.normalvariate(0, 100)

X = [random.normalvariate(10, 25) for _ in range(100)]
y = [f(x) for x in X]

plt.scatter(X, y)
plt.show()
```

Once we have our data, the next step is to "regress" X against Y.  This means we are saying there is a statistical relationship between X and Y.  Usually people take this to mean there is a causal relationship between X and Y, however this is not necessarily the case.  Statisticians usually mean, when X is statistically related to Y, that X covaries with Y in some statistical sense.

This may be because X and Y are derived from the same process or may have some hidden unexplained link.  Or it may just be that they both follow the same underlying functional form, without any semantic reasoning behind their covariance.
