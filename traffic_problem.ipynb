# Traffic Problem

We are city planners, deciding how to adjust an intersection.  We need to decide whether or not to make it wider or more narrow.

First let's look at the data:

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
%matplotlib inline
```

```python
num_rows = 24*7*4
one_hour = 3600 # seconds
traffic = np.random.normal(20, 10, num_rows)
df = pd.DataFrame(columns=["traffic", "timestamps"])
df["traffic"] = traffic
cur_hour = one_hour
df["timestamps"] = [datetime(2012,1,1,0,0,0)+timedelta(seconds=cur_hour + one_hour*elem) for elem in range(num_rows)]
```
